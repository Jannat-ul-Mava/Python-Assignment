```python
# Phone directory with phone numbers as keys and names as values
phone_directory = {
    '0348-6642185': 'Adeeb',
    '0321-4971961': 'Sohaib',
    '0321-5856463': 'Kamaal',
    '0321-4582322': 'Narmeen',
    '0345-8075177': 'Zuhra',
    '0323-7654511': 'Wajeeha'
}

# Sorting the directory by keys in ascending order
sorted_by_key_asc = dict(sorted(phone_directory.items()))
print("Sorted by key (ascending):", sorted_by_key_asc)

# Sorting the directory by keys in descending order
sorted_by_key_desc = dict(sorted(phone_directory.items(), reverse=True))
print("Sorted by key (descending):", sorted_by_key_desc)

# Sorting the directory by values in ascending order
sorted_by_value_asc = dict(sorted(phone_directory.items(), key=lambda item: item[1]))
print("Sorted by value (ascending):", sorted_by_value_asc)

# Sorting the directory by values in descending order
sorted_by_value_desc = dict(sorted(phone_directory.items(), key=lambda item: item[1], reverse=True))
print("Sorted by value (descending):", sorted_by_value_desc)
```
